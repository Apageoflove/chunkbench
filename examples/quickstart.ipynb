# ChunkBench Quickstart

Benchmark every RAG chunking strategy, not just guess.

```python
from chunkbench.benchmark import Benchmark

bench = Benchmark(
    documents="./examples/sample_docs/",
    qa_pairs="./examples/sample_qa.jsonl",
    strategies=["fixed_size", "recursive", "token_aware", "sentence_window"],
    embed_model="all-MiniLM-L6-v2",
    top_k=5,
    output_dir="./chunkbench_output",
)

report = bench.run()
```

```python
report.show_leaderboard()
```

```python
for insight in report.leaderboard.get_insights():
    print(f"- {insight}")
```

```python
report.plot_all()
```

```python
result = bench.run_single("recursive")
print(f"Chunks: {result.chunk_count}")
print(f"Avg size: {result.avg_chunk_size:.1f} chars")
print(f"Index time: {result.index_time:.2f}s")
```
